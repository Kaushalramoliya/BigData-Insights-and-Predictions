# Role of data engineer and data scientist

## Real-Time Stock Market Sentiment and Price Prediction

```python
import pandas as pd
import numpy as np
import yfinance as yf
import matplotlib.pyplot as plt
import statsmodels.api as sm
import time
import matplotlib

# In Google Colab, use inline plotting
%matplotlib inline

# --------------------------
# Define the top 10 Indian stocks (Yahoo Finance tickers on NSE have the ".NS" suffix)
# --------------------------
stocks = {
    "RELIANCE.NS": "Reliance Industries",
    "TCS.NS": "Tata Consultancy Services",
    "HDFCBANK.NS": "HDFC Bank",
    "INFY.NS": "Infosys",
    "ICICIBANK.NS": "ICICI Bank",
    "HINDUNILVR.NS": "Hindustan Unilever",
    "KOTAKBANK.NS": "Kotak Mahindra Bank",
    "LT.NS": "Larsen & Toubro",
    "BHARTIARTL.NS": "Bharti Airtel",
    "AXISBANK.NS": "Axis Bank"
}

def simulate_news_sentiment(dates):
    """
    Simulate daily sentiment data.
    Generates a random sentiment value between -1 and 1 for each date.
    """
    np.random.seed(42)  # For reproducibility; remove or modify for unique curves per stock if desired.
    sentiment = np.random.uniform(-1, 1, len(dates))
    sentiment_df = pd.DataFrame({'date': dates, 'sentiment': sentiment})
    return sentiment_df

def process_stock(ticker, company_name):
    print("\n" + "="*70)
    print(f"Processing {ticker} - {company_name}")
    print("="*70)

    # --------------------------
    # 1. Stock Data Collection
    # --------------------------
    stock_start = '2022-01-01'
    stock_end = '2023-12-31'
    stock_data = yf.download(ticker, start=stock_start, end=stock_end)
    if stock_data.empty:
        print(f"No stock data retrieved for {ticker}.")
        return

    # Flatten MultiIndex columns if present
    if isinstance(stock_data.columns, pd.MultiIndex):
        stock_data.columns = stock_data.columns.get_level_values(0)

    # Reset index if needed so that the date becomes a column
    if "Date" not in stock_data.columns:
        stock_data.reset_index(inplace=True)

    # Create a 'date' column in proper format.
    if "Date" in stock_data.columns:
        stock_data['date'] = pd.to_datetime(stock_data['Date']).dt.date
    else:
        stock_data['date'] = stock_data.index.date

    print("Stock Data Sample:")
    print(stock_data[['date', 'Close']].head())

    # --------------------------
    # 2. Simulate News Sentiment Data
    # --------------------------
    unique_dates = stock_data['date'].unique()
    sentiment_df = simulate_news_sentiment(unique_dates)
    print("Simulated News Sentiment Sample:")
    print(sentiment_df.head())

    # --------------------------
    # 3. Merge Stock Data with Simulated Sentiment
    # --------------------------
    merged_df = pd.merge(stock_data, sentiment_df, on='date', how='left')
    print("Merged Data Columns:", merged_df.columns.tolist())
    try:
        print("Merged Data Sample:")
        print(merged_df[['date', 'Close', 'sentiment']].head())
    except KeyError as e:
        print("KeyError during printing merged data:", e)
        return

    # --------------------------
    # 4. Feature Engineering (Optional)
    # --------------------------
    merged_df['MA_7'] = merged_df['Close'].rolling(window=7).mean()
    merged_df['daily_return'] = merged_df['Close'].pct_change()

    # --------------------------
    # 5. Data Visualization: Plot Closing Price and Sentiment Separately
    # --------------------------
    # Plot Closing Price
    plt.figure(figsize=(12, 6))
    plt.plot(merged_df['date'], merged_df['Close'], label='Closing Price', color='blue')
    plt.xlabel('Date')
    plt.ylabel('Closing Price')
    plt.title(f'{company_name} ({ticker}) - Closing Price (2022-2023)')
    plt.legend()
    plt.xticks(rotation=45)
    plt.tight_layout()
    price_plot_filename = f"{ticker}_closing_price.png"
    plt.savefig(price_plot_filename)
    print(f"Closing Price plot saved as {price_plot_filename}")
    plt.show()

    # Plot Simulated Sentiment
    plt.figure(figsize=(12, 6))
    plt.plot(merged_df['date'], merged_df['sentiment'], label='Simulated Sentiment', color='red')
    plt.xlabel('Date')
    plt.ylabel('Sentiment')
    plt.title(f'{company_name} ({ticker}) - Simulated Sentiment (2022-2023)')
    plt.legend()
    plt.xticks(rotation=45)
    plt.ylim(-1.2, 1.2)  # Set y-limits for clarity
    plt.tight_layout()
    sentiment_plot_filename = f"{ticker}_sentiment.png"
    plt.savefig(sentiment_plot_filename)
    print(f"Sentiment plot saved as {sentiment_plot_filename}")
    plt.show()

    # --------------------------
    # 6. SARIMAX Modeling & Forecasting
    # --------------------------
    ts_data = merged_df.set_index('date')
    close_prices = ts_data['Close']
    exog = ts_data['sentiment']

    try:
        model = sm.tsa.SARIMAX(close_prices, order=(1, 1, 1), exog=exog)
        model_fit = model.fit(disp=False)
        print("SARIMAX Model Summary (with simulated sentiment as exog):")
        print(model_fit.summary())

        # Forecast next 10 days. Assume future sentiment remains constant at last observed value.
        last_sentiment = exog.iloc[-1]
        future_exog = [last_sentiment] * 10
        forecast = model_fit.forecast(steps=10, exog=future_exog)
        print("\n10-Day Forecast of Closing Price:")
        print(forecast)
    except Exception as e:
        print(f"Error fitting SARIMAX model for {ticker}: {e}")

    # --------------------------
    # 7. Optional Model Evaluation (Train-Test Split)
    # --------------------------
    split_date = '2023-06-01'
    train_data = merged_df[merged_df['date'] < pd.to_datetime(split_date).date()]
    test_data = merged_df[merged_df['date'] >= pd.to_datetime(split_date).date()]

    if not train_data.empty and not test_data.empty:
        train_ts = train_data.set_index('date')
        test_ts = test_data.set_index('date')
        train_close = train_ts['Close']
        train_exog = train_ts['sentiment']
        test_close = test_ts['Close']
        test_exog = test_ts['sentiment']

        try:
            model_train = sm.tsa.SARIMAX(train_close, order=(1, 1, 1), exog=train_exog)
            model_train_fit = model_train.fit(disp=False)
            predictions = model_train_fit.forecast(steps=len(test_close), exog=test_exog)
            rmse = np.sqrt(np.mean((test_close - predictions) ** 2))
            mae = np.mean(np.abs(test_close - predictions))
            print(f"\nModel Evaluation for {ticker} on Test Data:")
            print(f"RMSE: {rmse}")
            print(f"MAE: {mae}")
        except Exception as e:
            print(f"Error during model evaluation for {ticker}: {e}")
    else:
        print("Not enough data for training/testing split for model evaluation.")

# --------------------------
# Main Loop: Process Each Stock
# --------------------------
for ticker, company in stocks.items():
    process_stock(ticker, company)
    time.sleep(5)
```
