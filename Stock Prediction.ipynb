```python
!pip install yfinance --quiet

import yfinance as yf
import pandas as pd
import matplotlib.pyplot as plt
import datetime

# Define a function to perform analysis for a given stock
def analyze_stock(ticker, event_time, event_name):
    """
    Analyzes stock data for a given ticker around an event.
    Plots four graphs:
      - 30 Minutes Before the event
      - 5 Minutes After the event
      - 1 Hour After the event
      - 1 Day After the event
    """

    # Define time windows relative to the event time
    # 30 Minutes After the event
    before_start = event_time - datetime.timedelta(minutes=30)
    before_end   = event_time

    # 5 Minutes After the event
    after_5m_start = event_time
    after_5m_end   = event_time + datetime.timedelta(minutes=5)

    # 1 Hour After the event
    after_1hr_start = event_time
    after_1hr_end   = event_time + datetime.timedelta(hours=1)

    # 1 Day After the event
    after_1day_start = event_time
    after_1day_end   = event_time + datetime.timedelta(days=1)

    # Data extraction function with error handling
    def get_stock_data(ticker, start, end, interval):
        data = yf.download(ticker, start=start, end=end, interval=interval, progress=False)
        if data.empty:
            print(f"No data available for {ticker} from {start} to {end} at interval {interval}.")
        return data

    # Fetch stock data for each window
    data_before    = get_stock_data(ticker, before_start, before_end, interval="1m")
    data_after_5m  = get_stock_data(ticker, after_5m_start, after_5m_end, interval="1m")
    data_after_1hr = get_stock_data(ticker, after_1hr_start, after_1hr_end, interval="5m")
    data_after_1day= get_stock_data(ticker, after_1day_start, after_1day_end, interval="60m")

    # Fetch news headlines using yfinance
    ticker_obj = yf.Ticker(ticker)
    news_list = ticker_obj.news  # List of dictionaries with news data

    def get_news_in_window(news_list, start, end):
        """Filter news items between start and end (datetime objects)."""
        news_in_window = []
        for news in news_list:
            pub_time = datetime.datetime.fromtimestamp(news.get('providerPublishTime', 0))
            if start <= pub_time <= end:
                headline = news.get('title', 'No Title')
                news_in_window.append((pub_time, headline))
        return news_in_window

    # Get news headlines for the post-event windows
    news_5m  = get_news_in_window(news_list, after_5m_start, after_5m_end)
    news_1hr = get_news_in_window(news_list, after_1hr_start, after_1hr_end)
    news_1day= get_news_in_window(news_list, after_1day_start, after_1day_end)

    # Plotting function for a given axis
    def plot_with_news(ax, data, title, news_events):
        """Plot stock data and annotate news events."""
        if not data.empty:
            ax.plot(data.index, data['Close'], marker='o', linestyle='-')
        else:
            ax.text(0.5, 0.5, 'No Data', horizontalalignment='center',
                    verticalalignment='center', transform=ax.transAxes)

        ax.set_title(title)
        ax.set_xlabel("Time")
        ax.set_ylabel("Close Price")

        # Annotate news events if available
        for event_time, headline in news_events:
            ax.axvline(event_time, color='red', linestyle='--', linewidth=1)
            ax.text(event_time, ax.get_ylim()[1]*0.95, headline[:50] + ('...' if len(headline)>50 else ''),
                    rotation=90, verticalalignment='top', fontsize=8, color='red')

    # Set up graph titles that include the event name
    title_before    = f"30 Minutes Before {event_name}"
    title_after_5m  = f"5 Minutes After {event_name}"
    title_after_1hr = f"1 Hour After {event_name}"
    title_after_1day= f"1 Day After {event_name}"

    # Create a figure with four subplots
    plt.figure(figsize=(14, 10))

    ax1 = plt.subplot(2, 2, 1)
    plot_with_news(ax1, data_before, title_before, news_events=[])

    ax2 = plt.subplot(2, 2, 2)
    plot_with_news(ax2, data_after_5m, title_after_5m, news_events=news_5m)

    ax3 = plt.subplot(2, 2, 3)
    plot_with_news(ax3, data_after_1hr, title_after_1hr, news_events=news_1hr)

    ax4 = plt.subplot(2, 2, 4)
    plot_with_news(ax4, data_after_1day, title_after_1day, news_events=news_1day)

    plt.tight_layout()
    plt.show()

    # Print a header for the analysis
    print(f"\n=== Analysis for {ticker} - {event_name} ===")

    print_news(news_5m, "5 Minutes After")
    print_news(news_1hr, "1 Hour After")
    print_news(news_1day, "1 Day After")

    # Optionally display the raw data for further analysis
    print("\nData Before Event (30 minutes prior):")
    display(data_before)

    print("\nData 5 Minutes After Event:")
    display(data_after_5m)

    print("\nData 1 Hour After Event:")
    display(data_after_1hr)

    print("\nData 1 Day After Event:")
    display(data_after_1day)

# -------------------------------------------------------------------
# Analysis for Apple Stock (AAPL) - iPhone 16 Launch
# -------------------------------------------------------------------
# Set event time for Apple (yesterday at 2:00 PM)
event_time_apple = (datetime.datetime.now() - datetime.timedelta(days=1)).replace(hour=14, minute=0, second=0, microsecond=0)
analyze_stock("AAPL", event_time_apple, "iPhone 16 Launch")

# -------------------------------------------------------------------
# Analysis for Google Stock (GOOGL) - Google Pixel Launch
# -------------------------------------------------------------------
# Set event time for Google (yesterday at 3:00 PM)
event_time_google = (datetime.datetime.now() - datetime.timedelta(days=1)).replace(hour=15, minute=0, second=0, microsecond=0)
analyze_stock("GOOGL", event_time_google, "Google Pixel Launch")
```
